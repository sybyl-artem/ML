```python
# импорт библиотек
import statsmodels.api as sm
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
from itertools import product
from datetime import *
from dateutil.relativedelta import *

%matplotlib inline
```

```python
# импорт данных из файла
data = pd.read_csv("WAG_C_M.csv", sep = ";", index_col = "month" ,parse_dates = True, dayfirst = True )
data.head()
```

#### 1. Визуальный анализ ряда

```python
# посмотрим на данные с помощью графика
plt.figure(figsize = (12,5))
data.WAG_C_M.plot()
plt.title("Zarplata")
```

Проверим стационарность ряда:

```python
stac = sm.tsa.seasonal_decompose(data.WAG_C_M).plot()
print ("Dickey-Fuller test, p = %f " % sm.tsa.stattools.adfuller(data.WAG_C_M)[1])
```

#### 2. Стабилизация дисперсии (при необходимости)
Воспользуемся преобразованием Бокса-Кокса

```python
plt.figure(figsize = (12,5))
data["WAG_C_M_cox"], lambda2 = stats.boxcox(data.WAG_C_M)
data.WAG_C_M_cox.plot()
plt.title("Zarplata_box_cox")
```

Выведем оптимальные параметры преобразования Бокса-Кокса и проведем тест Дикки-Фуллера

```python
print("Optimal parametr = %f" % lambda2)
print("Dickey-Fuller test result, p = %f" % sm.tsa.stattools.adfuller(data.WAG_C_M_cox)[1])
```

Согласно полученым параметрам, ряд является нестационарным. Но на графике все еще виден тренд, попробуем оптимизировать выборку.

#### 3. Выбор порядка дифференцирования

Воспольнуемся сезонным дифференцированием, поскольку ряд явно имеет сезонность.


```python
data["WAG_C_M_cox_diff"] = data.WAG_C_M_cox - data.WAG_C_M_cox.shift(12)
sm.tsa.seasonal_decompose(data.WAG_C_M_cox_diff.dropna()).plot()
print("Dickey-Fuller test result, p = %f" % sm.tsa.stattools.adfuller(data.WAG_C_M_cox_diff.dropna())[1])
```

В лекции было сказано, что проводить дифференциирование ряда можно несколько раз, пока тренд не пропадет с графика. Повторим процедуру, и сравним полученные результаты

```python
data["WAG_C_M_cox_diff2"] = data.WAG_C_M_cox_diff - data.WAG_C_M_cox_diff.shift(1)
sm.tsa.seasonal_decompose(data.WAG_C_M_cox_diff2.dropna()).plot()
print("Dickey-Fuller test result, p = %f" % sm.tsa.stattools.adfuller(data.WAG_C_M_cox_diff2.dropna())[1])
```

p = 0.000000 значение вообще близкое к нулю. Четкий тренд на графике отсутствует, по этому можно преступать к построению модели.

#### 4. Выбор начальных приближений для p, q, P, Q

Посмотрим на остатки при автокорреляции, построим соответствующие графики

```python
plt.figure(figsize = (12,7))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(data.WAG_C_M_cox_diff2.dropna()[12:].squeeze(), lags = 50, ax = ax)

plt.figure(figsize = (12,7))
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.WAG_C_M_cox_diff2.dropna()[12:].squeeze(), lags = 50, ax = ax)
```

```python
# Выберем следующие параметры приблежения:
ps = range(0,2)
d = 1
qs = range(0,2)
Ps = range(0,2)
D = 1
Qs = range(0,1)
```

```python
parameters = product(ps,qs,Ps, Qs)
parameters_list = list(parameters)
parameters_list
```

#### 5. Обучение и сравнение моделей-кандидатов, выбор победителя

```python
# обучим модели через цикл, с его помощью выберем лучшую.
results = []
best_aic = float("inf")

for param in parameters_list:
    model = sm.tsa.statespace.SARIMAX(data.WAG_C_M_cox, order = (param[0], d, param[1]), seasonal_order = (param[2], D, param[3],12)).fit(disp=-1)                                          
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# выведем результаты моделей, убедимся в их адекватности
pd.DataFrame(results)
```

```python
# выведем лучшую модель
print(best_model.summary())
```

#### 6. Анализ остатков построенной модели

```python
# построим графики остатков, применим тест Стьюдента и Дикки-Фулера.
plt.figure(figsize = (15,12))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel("Residuals")

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid.values.squeeze(), lags = 50, ax = ax)

print("Students test result: p = %f" % stats.ttest_1samp(best_model.resid[13:],0)[1])
print("Dickey-Fuller test result, p = %f " % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки получаются несмещёнными(Критерий Стьюдента), прогноз не нужно корректировать на константу; остатки стационарны(Критерий Дикки-Фуллера)  и неавтокоррелированы (коррелограмма, критерий Льюнга_Бокса). Соответственно, оптимально оставить эту модель для дальнейшей работы.

#### 7. Прогнозирование с помощью финальной модели.

Для начала, сравним на графике реальные данные и модельные.

```python
def boxcox2(y, lambda2):
    if lambda2 == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lambda2*y+1)/lambda2))

data["model"] = boxcox2(best_model.fittedvalues, lambda2)

plt.figure(figsize = (15,7))
data.WAG_C_M.plot()
data.model[13:].plot(color = "y")
```

Как видим, модель очень хорошо описывает исходный временной ряд.

```python
# строим непосредственно прогноз, на 36 месяцев, начиная с сентября 2016 года(последние доступные данные).
data2 = data[["WAG_C_M"]]
date_list = [datetime.strptime("2016-09-01", "%Y-%m-%d") + relativedelta(months = x) for x in range(0,36)]
future = pd.DataFrame(index = date_list, columns = data2.columns)
data2 = pd.concat([data2, future])
data2["forecast"] = invboxcox(best_model.predict(start = 284, end = 329), lambda2)

plt.figure(figsize = (12,5))
data2.WAG_C_M.plot()
data2.forecast.plot(color = "y")
plt.title("Salary forecast from 2016 to 2019")
```

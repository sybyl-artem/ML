```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("botswana.tsv", sep = "\t")
data["religion"].value_counts()
```

```python
data_na = data.dropna()
data_na.shape
```

```python
data["newermarr"] = [1 if data.loc[i,"evermarr"] == 0 else 0 for i in range (data.shape[0]) ]
data.agefm[data.agefm.isnull()] = 0
data.heduc[data.newermarr.values == 1 & data.heduc.isnull()] = -1
data.heduc.isnull().value_counts()
del data["evermarr"]
```

```python
data["idlnchld_noans"] = 0
data['heduc_noans'] = 0
data["usemeth_noans"] = 0


data.loc[data.idlnchld.isnull(), "idlnchld_noans"] = 1

data.loc[data.heduc.isnull(), "heduc_noans"] = 1

data.loc[data.usemeth.isnull(), "usemeth_noans"] = 1

```

```python
data.idlnchld[data.idlnchld.isnull()] = -1
data.heduc[data.heduc.isnull()] = -2
data.usemeth[data.usemeth.isnull()] = -1
```

```python
data
```

```python
data = data.dropna()
data.info()
```

```python
data.shape[0] * data.shape[1]
```

```python
import statsmodels.formula.api as smf

model = "ceb ~ " + "+".join(data.columns[1:])
reg_m = smf.ols(model, data)
fitted_model = reg_m.fit()
fitted_model.summary()
```

```python
import statsmodels.stats.api as sms
sms.het_breuschpagan(fitted_model.resid, fitted_model.model.exog)[1]
```

```python
model = "ceb ~ " + "+".join(data.columns[1:])
reg_m2 = smf.ols(model, data)
fitted_model2 = reg_m2.fit(cor_type = "HC1")
fitted_model2.summary()
```

```python
model = "ceb ~ " + "+".join(data.columns[1:])
model2 = 'ceb ~ age+educ+idlnchld+knowmeth+usemeth+agefm+heduc+urban+electric+bicycle+newermarr+idlnchld_noans+heduc_noans+usemeth_noans'
```

```python
reg_m3 = smf.ols(model2, data)
fitted_model3 = reg_m3.fit()
fitted_model3.summary()
```

```python
sms.het_breuschpagan(fitted_model3.resid, fitted_model3.model.exog)[1]
```

```python
reg_m4 = smf.ols(model2, data)
fitted_model4 = reg_m4.fit(cor_type = "HC1")
fitted_model4.summary()
```

```python
reg_m2.fit().compare_f_test(reg_m4.fit())
```

```python
model3 = 'ceb ~ age+educ+idlnchld+knowmeth+agefm+heduc+urban+electric+bicycle+newermarr+idlnchld_noans+heduc_noans'
reg_m5 = smf.ols(model3, data)
fitted_model5 = reg_m5.fit()
fitted_model5.summary()
```

```python
reg_m4.fit().compare_f_test(reg_m5.fit())
```
